```python
import pandas as pd

# I WON'T UPLOAD THE DATA TO OUR REPOSITORY.
# THE DATA CAN BE FOUND AT OUR ONEDRIVE A-TEAM DIRECTORY

# the delimeter in these files is a tab written as \t.
PATH_TO_FILES = "../data/"
```

```python
# Links company name with an uuid used within this domain
df_assignee = pd.read_csv(PATH_TO_FILES + "assignee.tsv", delimiter='\t', on_bad_lines='skip')

# All entries that are not linked to a company are deleted for a better overview.
# Type is also deleted because of insufficient information what it stands for.
df_assignee = df_assignee.drop(columns = ['name_first', 'name_last', 'type'])
df_assignee = df_assignee.dropna()

df_assignee
```

```python
names = ['Amazon.com, Inc.',
         'Amazon Technologies, Inc.',
         
         'APPLE INC.',
         'Apple Computer Inc.',
         
         'Tesla, Inc.',
         'Tesla Motors, Inc.',
         
         'Google LLC',
         'Alphabet, Inc.',
         
         'Microsoft Technology Licensing, LLC']

    ### Using loc is common practice, because of slicing vs viewing or something ###
    ### Do this before all the other stuff, so merging doesn't crash your pc ###
df_assignee_filtered = df_assignee.loc[df_assignee['organization'].isin(names)]
df_assignee_filtered = df_assignee_filtered.rename({'id': 'assignee_id'}, axis='columns')
df_assignee_filtered
```

```python
# Links company uuid with patent ids 
df_patent_assignee = pd.read_csv(PATH_TO_FILES + "patent_assignee.tsv", delimiter='\t', on_bad_lines='skip')

# All location uuids are also deleted because they are irrelevant for our analysis.
df_patent_assignee = df_patent_assignee.drop(columns = ['location_id'])

#df_patent_assignee = df_patent_assignee.astype({"patent_id": str})
#df_patent_assignee.patent_id = df_patent_assignee.patent_id.str.strip()
df_patent_assignee
```

```python
df_org_pat = df_assignee_filtered.merge(df_patent_assignee, on = 'assignee_id', how = 'inner')
df_org_pat.groupby('organization').size()
```

```python
df_org_pat = df_org_pat.replace('APPLE INC.','Apple')
df_org_pat = df_org_pat.replace('Apple Computer Inc.','Apple')

df_org_pat = df_org_pat.replace('Amazon Technologies, Inc.','Amazon.com')
df_org_pat = df_org_pat.replace('Amazon.com, Inc.','Amazon.com')

df_org_pat = df_org_pat.replace('Alphabet, Inc.','Google')
df_org_pat = df_org_pat.replace('Google LLC','Google')

df_org_pat = df_org_pat.replace('Microsoft Technology Licensing, LLC','Microsoft')

df_org_pat = df_org_pat.replace('Tesla Motors, Inc.','Tesla')
df_org_pat = df_org_pat.replace('Tesla, Inc.','Tesla')
```

```python
df_org_pat.groupby('organization').size()
```

```python
df_org_pat = df_org_pat.drop(columns = ['assignee_id'])
df_org_pat.head()
```

```python
df_sum = pd.DataFrame(columns={"uuid", "patent_id", "text", "year"})
for year in range(2015, 2020):
    temp_df = pd.read_csv(f"{PATH_TO_FILES}brf_sum_text_{year}.tsv", delimiter='\t', on_bad_lines='skip', dtype=str)
    temp_df['year'] = year
    df_sum = df_sum.append(temp_df)

df_sum = df_sum.astype({"patent_id": str})
df_sum.patent_id = df_sum.patent_id.str.strip()
df_sum
```

```python
df_overview = df_org_pat.merge(df_sum, on = 'patent_id', how = 'inner')
```
